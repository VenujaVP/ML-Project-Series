```python
import pandas as pd 
import matplotlib.pyplot as ptl
from sklearn.linear_model import LinearRegression  # Importing LinearRegression class from the sklearn library
```

```python
data = pd.read_csv('test.csv')
print(data)
```

```python
ptl.scatter(data['Area'] , data['Price'])  # this show scatter of Area & Price
ptl.plot(data['Area'] , data['Price'])  # this show line of Area & Price
ptl.show()
```

```python
# Creating an instance of the model
model = LinearRegression() 
```

```python
#train the model
model.fit(data[['Area']], data['Price'])  #data[['Area']] and data[['Price']] are DataFrame objects with one column each.
```

```python
#this show scatter of Area & model prediction of Ares(prices)
ptl.plot(data['Area'] , model.predict(data[['Area']]))
#this show scatter of Area & Price
ptl.scatter(data['Area'] , data['Price'])
```

```python
new_data = pd.read_csv('test new.csv')
print(new_data)

print(model.predict(new_data[['Area']]))
```
